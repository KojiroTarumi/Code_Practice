```python
# Index指定していないTableで, display(df.loc[1:3])と display(df.iloc[1:3]) が同じ結果にならない
# この問題に対し、iris data　setで検証してみる

import pandas as pd
from sklearn.datasets import load_iris
 
#アヤメのデータの詳細を出力 DESCR = describe(記述)
data = load_iris()
print(data.DESCR)
```

```python
# アヤメのデータ（説明変数）を X に格納する
X = pd.DataFrame(data=data.data,columns=data.feature_names)
X.head()
```

```python
# アヤメのデータ（目的変数）を y に格納する
y = pd.DataFrame(data=data.target)
y = y.rename(columns={0: 'Species'})
y.head()
```

```python
# Xとyを連結させる
# (参考) http://sinhrks.hatenablog.com/entry/2015/01/28/073327
# 横方向に連結したい場合は axis=1 を指定
df = pd.concat([X, y], axis=1)
df.head()
```

```python
# 各カラムのデータ型を表示する
df.dtypes
```

```python
# 有効データ数、データ型、メモリ使用量などの情報を表示
df.info()
```

```python
# 特徴量の平均値と標準偏差、四分位数を一度に表示
df.describe()
```

```python
# 各カラムの有効データ数を表示
df.count()
```

```python
# すべてのカラムに対して欠損値の数を計算
len(df) - df.count()
```

```python
# 特徴量に欠損値があるかを確認する
df.isnull().head()
```

#### .locや.ilocを使うことになる。この２つはどう違うのか

- loc()、iloc()、および ix() は、行と列を同時に指定することができる
- loc() はラベル名で指定 (index名、column名の順で指定する)
- iloc() は数値で指定 (index番号、column番号の順で指定)
- ix() ならラベル名と数値のいずれでも利用することができる

```python
# Index指定していないTableで, display(df.loc[1:3])と display(df.iloc[1:3]) が同じ結果にならない
# この問題に対し、iris data　setで検証してみる
```

```python
hoge = df.loc[1:3]
hoge
```

```python
# (参照) https://akiyoko.hatenablog.jp/entry/2017/04/03/081630#1-1

# (1)sepal_widthの列を抽出する - 1   　… df['カラム名'] この形式だとスペースを含むカラム名も使用可能
Sepal_width_1 = df[["sepal width (cm)"]]
Sepal_width_1.head()
```

```python
# (1)sepal_widthの列を抽出する - 2    … df.iloc[[行を指定, 列を指定]]
Sepal_width_2 = df.iloc[:, 2]    # [: …すべての行を指定, 2 …列を指定]
Sepal_width_2
```

```python
# (2) petal_length の列の50番目から99番目までのデータを抽出
Data_Extraction = df.loc[50:99, ['petal length (cm)']]
Data_Extraction
```
